```python
from datautils import load
df = load()
```

### Ensure Data is Clean via Great Expectations

```python
# expect columns to exist
for col in ['user_id', 'checkin', 'checkout', 'city_id', 'device_class', 'affiliate_id', 'booker_country', 'hotel_country', 'utrip_id']:
    df.expect_column_to_exist(col)

# No columns are null
for col in df.columns:
    df.expect_column_values_to_not_be_null(col)
    
# device_class in desktop/mobile/tablet
df.expect_column_values_to_be_in_set("device_class", ["desktop", "mobile", "tablet"])

# no duplicate rows
df.expect_compound_columns_to_be_unique(df.columns)

# checkout after checkin
df.expect_column_pair_values_A_to_be_greater_than_B('checkout', 'checkin')

# df.get_expectation_suite().to_json_dict()
df.validate()
```
